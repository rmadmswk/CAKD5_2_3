```python
from glob import glob
 
# 이미지들의 주소 리스트로 만들어줌
train_img_list = glob('/content/drive/MyDrive/project2/version9/yolov5/custom_data/train/images/*.jpg')
valid_img_list = glob('/content/drive/MyDrive/project2/version9/yolov5/custom_data/valid/images/*.jpg')
test_img_list = glob('/content/drive/MyDrive/project2/version9/yolov5/custom_data/test/images/*.jpg')

# 리스트를 txt파일로 생성
with open('/content/drive/MyDrive/project2/version9/yolov5/custom_data/train.txt', 'w') as f:
   f.write('\n'.join(train_img_list) + '\n')
    
with open('/content/drive/MyDrive/project2/version9/yolov5/custom_data/valid.txt', 'w') as f:
   f.write('\n'.join(valid_img_list) + '\n')

with open('/content/drive/MyDrive/project2/version9/yolov5/custom_data/test.txt', 'w') as f:
   f.write('\n'.join(test_img_list) + '\n')
```

```python
import yaml

with open('/content/drive/MyDrive/project2/version9/yolov5/custom_data/data.yaml', 'r') as f:
  data = yaml.full_load(f)

print(data)

data['train'] = '/content/drive/MyDrive/project2/version9/yolov5/custom_data/train.txt'
data['test'] = '/content/drive/MyDrive/project2/version9/yolov5/custom_data/test.txt'
data['val'] = '/content/drive/MyDrive/project2/version9/yolov5/custom_data/valid.txt'

with open('/content/drive/MyDrive/project2/version9/yolov5/custom_data/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
